# RAG Chain

```python
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "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"
repo_id = "mistralai/Mistral-7B-Instruct-v0.1"
llm = HuggingFaceHub(repo_id=repo_id, model_kwargs={"temperature": 0.1, "max_new_tokens": 1024})
```

```python
system_prompt = """You are a chatbot which helps the Muslilms with their queries of daily life.
        If you don't know the answer to a ny question from the documents provided to you, then apologize. Give maximu context of your answer.
        Given the Context and chat history answer the user questions."""
B_INST, E_INST = "<s>[INST] ", " [/INST]"
template = (
                B_INST
                + system_prompt
                + """

            Context: {chat_history} \n {context}
            User: {question}"""
                + E_INST
            )
prompt = PromptTemplate(input_variables=["context", "chat_history", "question"], template=template)
```

```python
memory = ConversationBufferMemory(memory_key='chat_history', return_messages=True, output_key='answer')
```

```python
retriever = db.as_retriever()
```

```python

chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory,
    return_generated_question=True,
    rephrase_question=True,
    return_source_documents=False,
    combine_docs_chain_kwargs={"prompt": prompt}
    )
```

```python
result=chain("A member of my family ordered a pizza from a takeaway which I do not trust as there is a rumour that the chicken could be dodgy, anyway one of the pizza slices touched my home cooked food which I was eating so my question is will this make my food haraam because I ate the roti which the slice touched?")
print(result['answer'])
```

```python
import gradio as gr
def chatllama(user_msg, history):
    result = chain(user_msg)
    return result['answer']

demo = gr.ChatInterface(fn=chatllama,)
demo.launch()
```
